```python
f1 = cv2.imread(r"C:\\Users\\oem\\Downloads\\flower.jpg")
f1 = cv2.cvtColor(f1,cv2.COLOR_BGR2RGB)
plt.imshow(f1)
plt.show()
```

#### we use kernels(filters) of different sizes to blur the image

blurring

```python
k1=np.ones((3,3))/9
k2=np.ones((5,5))/25
k3=np.ones((9,9))/81

r1=cv2.filter2D(f1,-1,k1)
r2=cv2.filter2D(f1,-1,k2)
r3=cv2.filter2D(f1,-1,k3)

plt.imshow(f1),plt.title('Original'),plt.show()
plt.imshow(r1),plt.title('Blurring with 3*3 Filter'),plt.show()
plt.imshow(r2),plt.title('Blurring with 5*5 Filter'),plt.show()
plt.imshow(r3),plt.title('Blurring with 9*9 Filter'),plt.show()

plt.show()
```

# sharpening , Embossing

```python
sharp = np.array([[-1,-1,-1],[-1,9,1],[-1,-1,-1]])   ##sharpening
emboss = np.array([[0,-1,-1],[1,0,-1],[1,1,1]])       ##embossing
```

```python
res1=cv2.filter2D(f1,-1,sharp)
res2=cv2.filter2D(f1,-1,emboss)

plt.imshow(f1),plt.title('Original'),plt.show()
plt.imshow(res1),plt.title('sharpening filter'),plt.show()
plt.imshow(res2),plt.title('embossing filter'),plt.show()
plt.show()
```
